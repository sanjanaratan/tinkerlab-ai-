```python
# Dummy tweets to simulate scraped results
tweets = [
    "Zomato's delivery time has really improved lately!",
    "I had such a terrible experience with Zomato last night.",
    "Zomato is okay, not too great not too bad.",
    "Absolutely love the new offers from Zomato!",
    "Zomato's UI is kind of confusing lately.",
    "Honestly, I rely on Zomato all the time. Great service!"
]
```

```python
!pip install certifi
import certifi
import os
os.environ['SSL_CERT_FILE'] = certifi.where()
```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

for i, tweet in enumerate(tweets):
    result = classifier(tweet[:512])[0]
    print(f"{i+1}. {tweet}")
    print("Sentiment:", result['label'], "| Confidence:", round(result['score'], 2))
    print()
```

```python
results = []

for tweet in tweets:
    result = classifier(tweet[:512])[0]
    sentiment = result['label']
    score = round(result['score'], 2)
    results.append((tweet, sentiment, score))

# Display in a simple table
from tabulate import tabulate

print(tabulate(results, headers=["Tweet", "Sentiment", "Confidence"], tablefmt="grid"))
```

```python
from transformers import pipeline

# Load HuggingFace sentiment model
classifier = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# AI assistant-style wrapper
def ai_sentiment_bot(tweet):
    print("Analyzing sentiment for this tweet:")
    print(f" {tweet}")
    result = classifier(tweet[:512])[0]
    print(f" AI says: '{result['label']}' with confidence {round(result['score'], 2)}")

# Try it
ai_sentiment_bot("Zomato is really improving its delivery time lately!")
```
